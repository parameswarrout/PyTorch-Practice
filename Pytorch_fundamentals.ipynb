# Squeeze, Unsqueeze, Permute

```python
x = torch.arange(1., 10.)
x, x.shape
```

```python
x_reshaped = x.reshape(1, 9)
x_reshaped, x_reshaped.shape
```

```python
x_reshaped = x_reshaped.squeeze()
x_reshaped, x_reshaped.shape
```

```python
x_reshaped = x_reshaped.unsqueeze(dim=0)
x_reshaped, x_reshaped.shape
```

```python
x_reshaped = x_reshaped.unsqueeze(dim=3)
x_reshaped, x_reshaped.shape
```

```python
x_reshaped = x_reshaped.squeeze(dim=0)
x_reshaped, x_reshaped.shape
```

```python
x_reshaped = x_reshaped.squeeze(dim=2)
x_reshaped, x_reshaped.shape
```

```python
x_original = torch.rand(size=(224, 224, 3))

x_permutated = x_original.permute(2, 0, 1)

print(f"Previous shape: {x_original.shape}")
print(f"New Shape: {x_permutated.shape}")
```

# Putting tensors (and models) on the GPU

```python
tensor = torch.tensor([1, 2, 3])

print(tensor, tensor.device)
```

```python
# Move tensor to GPU (if available)
tensor_on_gpu = tensor.to(device)
tensor_on_gpu
```

# Moving tensor back to the CPU

```python
tensor_on_gpu.numpy()
```

```python
# To fix the GPU tensor with numpy issue, we can first set it to the CPU
tensor_back_on_cpu = tensor_on_gpu.cpu().numpy()
tensor_back_on_cpu
```
